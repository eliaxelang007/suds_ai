```python
from ragas import evaluate
from ragas.metrics import (
    faithfulness,
    answer_relevancy,
    answer_correctness,
    context_recall,
    context_precision,
)

from datasets import Dataset
from pandas import read_csv

from ast import literal_eval

def main():
	metrics = [
		faithfulness,
		answer_relevancy,
		context_recall,
		context_precision,
		answer_correctness,
	]

	pandas_response_dataset = read_csv("response_set.csv", converters={"contexts": literal_eval})
	response_dataset = Dataset.from_pandas(pandas_response_dataset)

	evaluation = evaluate(
		response_dataset,
		metrics
	)

	return evaluation.to_pandas()

result = None

if __name__ == "__main__":
	result = main()

result
```
